# Breakthrough plotting notebook

This notebook shows how to use the Plotly breakthrough tools in Jupyter.

It covers:
- loading metadata from `simulation.json`
- plotting breakthrough curves in time and dimensionless time
- using the interactive column explorer widget

Expected files in the working directory:
- `simulation.json`
- `column.data`
- `component_*.data`


```python
from ruptura.plot_breakthrough import BreakthroughPlotly, ColumnDataExplorer
```

## Build the plotter from `simulation.json`

Set `carrierGasComponent` to the component index of the carrier gas if you want it identified and optionally emphasized in the plots.

```python
plotter = BreakthroughPlotly.from_simulation_json(
    'simulation.json',
    data_dir='.',
    carrierGasComponent=None,
)
```

## Breakthrough curve in dimensionless time

```python
fig = plotter.breakthrough_dimensionless(include_carrier_gas=True)
fig
```

## Breakthrough curve in physical time

```python
fig = plotter.breakthrough_time(include_carrier_gas=True)
fig
```

## Column animation

Use `every` to subsample frames when a run stores many blocks.

```python
fig = plotter.column_animation('P', every=5)
fig
```

## Interactive widget explorer

The widget is useful when you want to inspect many components and many stored blocks quickly.

How the controls work:
- `x:` chooses between axial position (`grid (z)`) and time.
- `y:` selects the stored quantity, such as `Q`, `Qeq`, `P`, `Pnorm`, `Dpdt`, or `Dqdt`.
- `block` is used in grid mode and selects one stored profile from `column.data`.
- `z@` is used in time mode and picks the nearest grid position to track through time.
- `components` lets you include or exclude traces interactively.

Grid mode answers: *what does the column profile look like at one stored time?*

Time mode answers: *how does one location in the column evolve over time?*

```python
explorer = plotter.explorer(
    datafile='column.data',
    dt=None,
)
explorer.display()
```
