# Mixture prediction plotting notebook

This notebook shows how to use the Plotly mixture-prediction tools in Jupyter.

It covers:
- loading metadata from `simulation.json`
- plotting pure-component isotherms
- plotting mixture loadings
- plotting adsorbed-phase mole fractions

Expected files in the working directory:
- `simulation.json`
- `component_*.data`


```python
from ruptura.plot_mixture import MixturePredictionPlotly
```

## Build the plotter from `simulation.json`

```python
plotter = MixturePredictionPlotly.from_simulation_json(
    'simulation.json',
    data_dir='.',
    carrierGasComponent=None,
)
plotter
```

## Pure-component isotherms

```python
fig = plotter.pure_components(include_carrier_gas=True)
fig
```

## Mixture loadings

```python
fig = plotter.mixture_loading(include_carrier_gas=True)
fig
```

## Adsorbed-phase mole fractions

```python
fig = plotter.mixture_adsorbed_molfractions(include_carrier_gas=True)
fig
```

## Excluding the carrier gas

For many systems this is the cleaner scientific view, since the carrier gas may have negligible uptake and can compress the visual scale for the adsorbing species.

```python
fig = plotter.mixture_loading(include_carrier_gas=False)
fig
```
