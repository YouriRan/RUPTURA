```python
import numpy as np
import matplotlib.pyplot as plt
import ruptura
```

```python
components = ruptura.Components()
components.addComponent(
    name="nC6",
    gasPhaseMolFraction=0.2,
    isotherms=[["Langmuir-Freundlich", 1.1, 2.14e-4, 0.6], ["Langmuir-Freundlich", 4.8, 2.39e-5, 1.35]]
)
components.addComponent(
    name="2MP",
    gasPhaseMolFraction=0.2,
    isotherms=[["Langmuir-Freundlich", 1.4, 1.68e-4, 0.6], ["Langmuir-Freundlich", 4.8, 3.3e-5, 1.26]]
)
components.addComponent(
    name="3MP",
    gasPhaseMolFraction=0.2,
    isotherms=[["Langmuir-Freundlich", 1.75, 1.14e-4, 0.62], ["Langmuir-Freundlich", 4.8, 3.22e-5, 1.24]]
)
components.addComponent(
    name="23DMB",
    gasPhaseMolFraction=0.2,
    isotherms=[["Langmuir-Freundlich", 1.5, 8.31e-5, 0.63], ["Langmuir-Freundlich", 4.8, 5.46e-5, 1.13]]
)
components.addComponent(
    name="22DMB",
    gasPhaseMolFraction=0.2,
    isotherms=[["Langmuir-Freundlich", 1.74, 5.65e-5, 0.72], ["Langmuir-Freundlich", 4.8, 4.3e-5, 1.04]]
)

mix = ruptura.MixturePrediction(
    displayName="CoBDP",
    temperature=443.0, 
    pressureStart=1e0,
    pressureEnd=1e8,
    numberOfPressurePoints=100,
    pressureScale="log",
    components=components
)
```

```python
mix.components.labels
```

```python
fig, ax = plt.subplots(3, 2, sharex=True, figsize=(12, 12))

mix.compute()
mix.plot(ax[0,0], "pure")
mix.plot(ax[1,0], "mixture")
mix.plot(ax[2,0], "mixture_molfrac")

params = mix.getComponentsParameters()
print(params)
params[1] *= 100
params[7] *= 100
params[-6] *= 2
params[-5] *= 100


print(params)
mix.setComponentsParameters(params)
mix.compute()

mix.plot(ax[0,1], "pure")
mix.plot(ax[1,1], "mixture")
mix.plot(ax[2,1], "mixture_molfrac")

fig.tight_layout()
```
